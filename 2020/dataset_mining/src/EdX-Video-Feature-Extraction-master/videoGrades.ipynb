```python
import os
import json
import csv

courseStructure = []       
with open('/Users/davidlemay/Google Drive/MOOC Time Series Analysis/Bucket 1/TeachersCollegeX-BDE1x-2T2015-course_structure-prod-analytics.json', 'r') as f:
    courseStructure = json.load(f)
    
for item in courseStructure.items():
    if item[1]['category'] == 'chapter': #chapters are organized into sequentials
        print(item[1]['metadata'].get('display_name'))
        sequentials = item[1]['children']
        for sequence in sequentials: #sequentials are organized into verticals
            verticals = courseStructure[sequence]['children'] 
            for vertical in verticals: #verticals are groups of content
                contents = courseStructure[vertical]['children']
                for content in contents:
                    print(courseStructure[content].get('category')+': '+content)
                    if courseStructure[content]['metadata'].get('display_name') is not None:
                        print('\tName: '+courseStructure[content]['metadata']['display_name'])
        print('\n')
```
